```python
# Imports
import os
from radon.complexity import cc_visit, cc_rank
import pandas as pd
```

```python
# Define the path to the folder containing the algorithms
path = 'Algorithms'
```

```python
# Function to analyze the complexity of a Python file
def analyze_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        code = file.read()
    return cc_visit(code)
```

```python
import pandas as pd

# Loop through each Python file
for filename in os.listdir(path):
    if filename.endswith('.py'):
        print(f"\n {filename}")
        results = analyze_file(os.path.join(path, filename))

        # Prepare data for DataFrame
        data = []
        for r in results:
            data.append({
                "Name": r.name,
                "Line": r.lineno,
                "Complexity": r.complexity,
                "Rank": cc_rank(r.complexity),
                "Type": r.__class__.__name__
            })

        # Create and display table
        df = pd.DataFrame(data)
        if not df.empty:
            display(df.sort_values(by="Complexity", ascending=False).reset_index(drop=True))
        else:
            print("No functions or methods detected.")
```
